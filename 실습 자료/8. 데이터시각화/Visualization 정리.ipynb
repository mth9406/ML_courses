## 1. plt.subplots

```python
fig, axs = plt.subplots(nrows= 3, ncols= 3, figsize=(15,10)) 
# figsize 인자를 이용해서 출력 크기를 조절할 수 있습니다.
# fig: 그림을 그릴 종이로 이해하면 쉽습니다.
# axes: "종이"에 그릴 xy좌표들로 이해하면 쉽습니다. 아래 결과 출력에서 네모들입니다. 
# (3,3) 으로 설정했기 때문에 총 3 * 3 배열로 axes 가 생성된 것을 확인 할 수 있습니다. 

axs[0,0].set_facecolor("#ddddff") # 각 축에 대한 접근은 numpy indexing 과 동일합니다. 
axs[2,2].set_facecolor("#ffffdd")

plt.tight_layout() 
# 축 간 거리를 조정합니다. 
# 왜 필요한지는 없어보면 압니다.
plt.show()
```

## 2. plt.gridspec.Gridspec

Gridspec 을 사용하면 여러 개의 그래프를 그릴 때 좌표축의 모양을 좀 더 정교하게 셋팅할 수 있습니다.

```python
from matplotlib.gridspec import GridSpec
```

```python
fig = plt.figure() # 그림을 그릴 종이를 가져옵니다.  
fig.suptitle("Controlling spacing around and between subplots") # 종이에 제목을 적습니다.
 
gs1 = GridSpec(3, 3) 
# 3*3 gridspc을 만듭니다. 
# 이제 gs1의 "인덱싱"을 이용해서 subplots()에서 봤던 네모의 크기를 조절할 수 있습니다.
# 예시를 보면서 설명을 드리겠습니다. 

ax1 = fig.add_subplot(gs1[:-1, :]) # fig.add_subplot()을 이용해서 xy 좌표를 추가할 수 있습니다. 
ax2 = fig.add_subplot(gs1[-1, :-1]) 
ax3 = fig.add_subplot(gs1[-1, -1])

plt.tight_layout()
plt.show()
```

## 3. axes 셋팅

본 챕터에서는 axes에 그래프를 그리는 방법에 대해서 알아보겠습니다.    
먼저 위에서 배운 내용을 잊지 않기 위해 다음 fig, axes 를 계속해서 사용하도록 합니다.

```python
fig = plt.figure(figsize= (15,10)) # 그림을 그릴 종이를 가져옵니다.  
fig.suptitle("Controlling spacing around and between subplots") # 종이에 제목을 적습니다.
 
gs1 = GridSpec(3, 3) 
# 3*3 gridspc을 만듭니다. 

ax1 = fig.add_subplot(gs1[:-1, :]) # fig.add_subplot()을 이용해서 xy 좌표를 추가할 수 있습니다. 
ax2 = fig.add_subplot(gs1[-1, :-1]) 
ax3 = fig.add_subplot(gs1[-1, -1])

plt.tight_layout()
plt.show()
```

### (1) ax1 에 sin 그래프를 그려보도록 하겠습니다.   
그래프를 그리기 위해 sin 함수에 대한 관측치를 가져옵니다.

```python
import numpy as np
x = np.linspace(start= 0, stop= np.pi, num= 1000) # 구간 [0,1) 을 1000 등분합니다.
y = np.sin(x) # sin 함수
```

이제 sin 함수를 ax1에 플랏팅하겠습니다.

```python
# 이제 sin 함수를 ax1에 플랏팅하겠습니다.
ax1.plot(x,y)
fig
```

0.5 간격인 major locator 사이에 0.1 간격의 minor locator를 삽입하겠습니다.

```python
# 축 간격 조절에는 다음을 사용합니다. 
from matplotlib.ticker import MultipleLocator as ML
from matplotlib.ticker import ScalarFormatter as SF
```

```python
ax1.xaxis.set_minor_locator(ML(0.1)) # 0.1 간격 minor locator를 삽입합니다.
ax1.xaxis.set_minor_formatter(SF()) # Format tick values as a number.
```

```python
fig
```

major 축 간격을 1.0 단위로 바꿔보겠습니다.

```python
ax1.xaxis.set_major_locator(ML(1.0)) # major locator 의 간격을 1.0 으로 셋팅합니다.
```

```python
fig
```

### (2) ax2 에 sin 함수를 그리되 x축의 scale 을 log로 셋팅하겠습니다.    
ax2.set_xscale() 사용

```python
ax2.set_xscale("log")
ax2.plot(x,y)
fig
```

ax2 에는 cos 함수도 같이 그려보겠습니다.    
하는 방법은 쉽습니다. ax2.plot() 에 코사인 함수를 넣어주면 됩니다.     
이 전에 그렸던 그림(사인 함수)은 지워지지 않기 때문입니다.

```python
# 먼저 코사인 함수 관측치를 y2 변수에 저장하겠습니다. 
# x 는 같은 값을 사용합니다.
y2 = np.cos(x)
```

```python
ax2.plot(x,y2)
fig
```

ax2 에 막상 플랏팅을 해보니 누가 cos 이고 누가 sin 인지 알았으면 좋겠습니다.    
이 경우 legend 를 사용합니다. 

```python
# 함수를 그린 순서대로 label을 달면 됩니다.
ax2.legend(['sin', 'cos'])
```

```python
fig
```

```python
# 만약 ax2 에서 cos 그래프를 지우고 싶다면 del 을 사용하시면 됩니다.
# ax2 가 갖고 있는 line 확인
ax2.lines
```

```python
# 두 번째 instance가 cos 함수 이므로
del ax2.lines[1]  
```

```python
# cos 함수가 지워진 것 확인
fig
```

```python
# legend 도 다시 설정해줍니다.
ax2.legend(['sin'])
```

```python
fig
```

```python
# y_lim() 함수를 이용해서 y 축 의 범위를 설정할 수 있습니다. 
# 이경우 0으로 두겠습니다. 
ax2.set_ylim(bottom= 0)
```

```python
fig
```

### (3) 이번에는 ax3 에 2개의 함수를 동시에 그리되 처음부터 label을 달아보겠습니다.

```python
ax3.plot(x, y, label= 'sin')
ax3.plot(x, y2, label= 'cos')
ax3.legend() # legend() 함수를 호출해야 위에서 설정한 label이 그려집니다.
```

```python
fig
```

## 3. axes 에는 다양한 종류의 그래프를 그릴 수 있습니다.  
subplots 에 다양한 그래프를 그려보겠습니다.

```python
fig, axes = plt.subplots(3,3, figsize = (15,10))
```

```python
# scatter plot
x, y = np.random.uniform(0, 1, 100), np.random.uniform(0, 1, 100) # uniform distribution 에서 100개 랜덤 샘플링
axes[0,0].scatter(x,y)
fig
```

```python
# bar 그래프
X = np.arange(10)
Y = np.random.uniform(1, 10, 10)
axes[0,1].bar(X, Y)
fig
```

```python
x = np.linspace(0,2*np.pi, 100)
y = np.sin(x)
axes[0,2].plot(x,y, label= 'sin')
axes[0,2].legend()
fig
```

```python
# image 
Z = np.random.uniform(0, 1, (8,8)) # 8*8 dimension.
axes[1,0].imshow(Z)
fig
```

```python
# Contour
Z = np.random.uniform(0, 1, (8,8))
axes[1,1].contourf(Z)
fig
```

```python
# 파이 차트
Z = np.random.uniform(0, 1, 4)
axes[1,2].pie(Z)
fig
```

```python
# 히스토그램
Z = np.random.normal(0, 1, 100)
axes[2,0].hist(Z)
fig
```

```python
# error bar 그래프
X = np.arange(5)
Y = np.random.uniform(0, 1, 5)
yerr= Y/4 # 암의로 각 관측치의 1/4 만큼 error 가 있다고 하겠습니다.
```

```python
axes[2,1].errorbar(x=X,y=Y, yerr=yerr, xerr=None) # yerr과 xerrr 은 default 가 None 입니다.
fig
```

```python
# 박스 플롯
Z = np.random.normal(loc=0, scale= 1, size = (100,3)) 
axes[2,2].boxplot(Z) # 3개의 컬럼 별로 값의 분포를 나타냅니다. 
fig
```
